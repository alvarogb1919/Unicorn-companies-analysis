```python
def obtain_funding(u):
    response = requests.get(u)
    soup = BeautifulSoup(response.text, 'html.parser')
    company_funding = (soup.find_all(name="span", class_="tooltip-target")[0].text)
    return company_funding
```

```python
obtain_funding("https://app.dealroom.co/companies/yuanfudao")
```

```python
def funding_normalizer(x):
    if "b" in x:
        y = x.replace("b", "")
        if "â‚¬" not in y:
            return("Not available")
        else:
            z = y.replace("â‚¬", "")
            return float(z)*1000
    elif "m" in x:
        y = x.replace("â‚¬", "")
        z = y.replace("m", "")
        return float(z)
    elif "k" in x:
        y = x.replace("â‚¬", "")
        z = y.replace("k", "")
        return float(z)/1000
    else:
        return("Not available")
```
